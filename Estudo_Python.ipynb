# **⏭ Usando herança multipla e polimorfismo**







```python
class Familia:
      def __init__(self, nome_completo):
          self.nome_completo = nome_completo

      def info_nome_completo(self):
          print(f"Meu nome é {self.nome_completo}")
```

```python
class Parentesco:
      def __init__(self, parentesco):
            self.parentesco = parentesco

      def info_parentesco(self):
            print(f"Meu grau de parentesco é: {self.parentesco}")
```

```python
class Endereco:
      def __init__(self, endereco_completo):
            self.endereco_completo = endereco_completo

      def info_endereco_completo(self):
            print(f"Meu endereço é: {self.endereco_completo}")

```

```python
class DadosFamiliar(Familia, Parentesco, Endereco):
      def __init__(self, nome_completo, parentesco, endereco_completo):
        Familia.__init__(self, nome_completo)
        Parentesco.__init__(self, parentesco)
        Endereco.__init__(self, endereco_completo)

      # Sobrescrevendo o metodo da classe familia
      def  info_nome_completo(self):
           print(f"Porra, ja falei que meu nomé é {self.nome_completo}")

```

```python
dados = DadosFamiliar("Ubirajara Moura da Silva",
                      "Pai",
                      "Rua Ignácio Florêncio da SIlveia, 108"
                      )


dados.info_parentesco()
dados.info_endereco_completo()
dados.info_nome_completo()
#
```
